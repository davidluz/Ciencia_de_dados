4 - Faça uma análise exploratória com as medidas vistas em aula (média, mediana, moda, obliquidade e kurtose) nos atributos anteriormente selecionados. Explique quais medidas ajudam no entendimento do conjunto de dados. Sinta-se livre para utilizar outras medidas, mas explique porque você está usando elas. Se achar necessário, pode utilizar mais atributos na análise.

```python
import pandas as pd
import scipy.stats as stats

# Selecionar os atributos relevantes
attributes = ['NUM_SALAS_EXISTENTES', 'NUM_FUNCIONARIOS', 'MATTOTAL']

# Criar um DataFrame com os atributos selecionados
df_selected = df[attributes]

# Calcular a média, mediana e moda para cada atributo
for attribute in attributes:
    mean = df_selected[attribute].mean()
    median = df_selected[attribute].median()
    mode = df_selected[attribute].mode().values[0]
    skewness = stats.skew(df_selected[attribute].dropna())
    kurtosis = stats.kurtosis(df_selected[attribute].dropna())

    print(f'Atributo: {attribute}')
    print(f'Média: {mean}')
    print(f'Mediana: {median}')
    print(f'Moda: {mode}')
    print(f'Obliquidade (Skewness): {skewness}')
    print(f'Kurtose: {kurtosis}')
    print('---'*20)
```

Com relação a quantidade de salas, a média e mediana sugerem que, em média, as escolas têm cerca de 10 salas, mas a distribuição é assimétrica à direita, indicando que há escolas com um número significativamente maior de salas. A obliquidade (skewness) positiva e a alta kurtose mostram que a distribuição é bastante inclinada à direita e tem caudas mais pesadas em comparação com uma distribuição normal. Isso indica a presença de outliers com um número muito maior de salas.

Olhando para matrículas, em média, as escolas têm cerca de 224 matrículas. A obliquidade (skewness) positiva e a kurtose relativamente baixa indicam que a distribuição é inclinada à direita, mas não tanto quanto os outros dois atributos.

Essa análise pode ajudar a identificar as escolas que precisam de mais recursos ou que estão superlotadas, bem como aquelas que podem ser consideradas atípicas em comparação com a maioria das escolas.

```python
import seaborn as sns

# Selecionar os atributos relevantes
attributes = ['NUM_SALAS_EXISTENTES', 'NUM_FUNCIONARIOS', 'MATTOTAL']

# Criar um DataFrame com os atributos selecionados
df_selected = df[attributes]

# Configurar o layout dos gráficos
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Criar o boxplot para cada atributo
for index, attribute in enumerate(attributes):
    # Selecionar o eixo apropriado
    ax = axes[index]

    # Criar o boxplot
    sns.boxplot(data=df_selected, y=attribute, ax=ax)

    # Configurar os rótulos do eixo y
    ax.set_ylabel(attribute)

# Mostrar os boxplots
plt.tight_layout()
plt.show()
```

7 - Detectar se existem outliers. Se existirem, quais são eles?

```python
attributes = ["NUM_SALAS_EXISTENTES", "NUM_FUNCIONARIOS", "MATTOTAL"]

for attribute in attributes:
    Q1 = df[attribute].quantile(0.25)
    Q3 = df[attribute].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[attribute] < lower_bound) | (df[attribute] > upper_bound)]

    print(f"Atributo: {attribute}")
    print(f"Outliers encontrados: {len(outliers)}")
    print(outliers)
    print("------------------------------------------------------------")
```
